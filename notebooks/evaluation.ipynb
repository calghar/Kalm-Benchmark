```python
from pathlib import Path

import pandas as pd
from kalm_benchmark.evaluation.evaluation import load_benchmark, Col,create_evaluation_summary
```

```python
import os
os.chdir("..")  # go from `notebooks` folder to the root folder


DATA_DIR = Path("./data").absolute()
DATA_DIR
```

```python

df = create_evaluation_summary(data_dir=DATA_DIR, show_extra=False)
```

```python
df
```

```python
cat_columns = ["cat_IAM", "cat_network", "cat_admission_ctrl", "cat_data_security", "cat_workload", "custom_checks"]
```

```python
publ_columns=[Col.Name, "version", "score", "coverage"] + cat_columns
```

```python
def rename_column(col_name: str) -> str:
    return col_name.replace("cat_", "").replace("_", " ").title()
```

```python
df_summary = df[publ_columns].rename(rename_column, axis=1)
```

```python
df_summary = df_summary.style.format({
    "Score": '{:,.2f}'.format,
    "Coverage": '{:,.1%}'.format,
    "Custom Checks": lambda x: "" if x=="False" else '\checkmark'
}, escape="latex").hide(axis="index")
```

```python
tex = df_summary.to_latex(caption="Summary of all evaluated open-source tools", label="tbl:tool-eval-summary",position_float="centering", hrules=True, column_format="ll|rr" + 'r' * len(cat_columns))
tex = tex.replace("table", "table*")  # make it span multiple columns
tex = tex.replace("%", "\\%")  # escape % so it can be compiled in LaTex
#tex = tex.replace("{llrr", "{ll|rr")  # add vertical line after version, to separate tool and eval results
```
